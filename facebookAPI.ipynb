## 以下示範實際操作的結果

```r
library(Rfacebook)
library(tidyverse)

## 取得 authenication, 要填入實際的secrets...
app_secret = "XXX"
app_id = "XXX"
fb_oauth <- fbOAuth(app_id, app_secret, extended_permissions = TRUE)
```

```r
#截取部分的「風生活」公開貼文 (n = 100)
all_posts <- getPage("SMediaLife", 
                     token = fb_oauth, n = 100, 
                     reactions = F)
all_posts %>% head
```

```r
#以其中一篇文為例，去截取使用者的 reactions, 數量不限(n=Inf)
post_id = "1510698909207091_1954039471539697"
this_post <- getPost(post_id, token = fb_oauth, n = Inf , reactions = T)
```

```r
this_post$reactions %>% head
```

```r
this_post$comments %>% head(n=10)
```

## 觀看 getPage(), getPost() 背後的 implementation
可以觀察到：
1. getPage用的是 https://graph.facebook.com/{page_id}/posts?fields={...} 這個endpoint
2. getPost用的是 https://graph.facebook.com/{post_id}?fields={...} 這個endpoint

細節需要查閱fb graph API reference
1. https://developers.facebook.com/docs/graph-api/reference
2. https://developers.facebook.com/docs/graph-api/reference/page
3. https://developers.facebook.com/docs/graph-api/reference/v2.10/post
